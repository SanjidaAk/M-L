```python
def lemmatize_text(text):
    lemmatizer = WordNetLemmatizer()
    tokens = word_tokenize(text)
    lemmatized_tokens = [lemmatizer.lemmatize(token, get_wordnet_pos(token)) for token in tokens]
    lemmatized_text = ' '.join(lemmatized_tokens)
    return lemmatized_text

def get_wordnet_pos(token):
    tag = nltk.pos_tag([token])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ, "N": wordnet.NOUN, "V": wordnet.VERB, "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

def remove_articles_and_auxiliary_verbs(sentence):
    words = sentence.split()
    filtered_words = [word for word in words if word.lower() not in ['a', 'an', 'the', 'is', 'am', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'do', 'does', 'did','of','very','to','and','one']]
    result = ' '.join(filtered_words)
    return result

def stem_paragraph(paragraph):
    stemmer = PorterStemmer()
    tokens = word_tokenize(paragraph)
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    stemmed_text = ' '.join(stemmed_tokens)
    return stemmed_text

```

```python
def preprocessed(para):
  rem_art =  remove_articles_and_auxiliary_verbs(para)
  stem = stem_paragraph(rem_art)
  lemma = lemmatize_text(stem)

  return lemma
```

```python
def calculate_jaccard_similarity(str1, str2):
    # Tokenize the strings and remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens1 = [word.lower() for word in word_tokenize(str1) if word.isalnum() and word.lower() not in stop_words]
    tokens2 = [word.lower() for word in word_tokenize(str2) if word.isalnum() and word.lower() not in stop_words]


    intersection = len(set(tokens1).intersection(tokens2))
    union = len(set(tokens1).union(tokens2))

    jaccard_similarity = intersection / union if union != 0 else 0
    return jaccard_similarity

def calculate_hamming_distance(str1, str2):

    max_length = max(len(str1), len(str2))
    min_length = min(len(str1), len(str2))

    str1 = str1.ljust(max_length)
    str2 = str2.ljust(max_length)

    differing_chars = sum(c1 != c2 for c1, c2 in zip(str1, str2))

    hamming_distance = differing_chars / max_length
    return hamming_distance

def preprocess_text(text):
    stop_words = set(stopwords.words('english'))
    tokens = [word.lower() for word in word_tokenize(text) if word.isalnum() and word.lower() not in stop_words]
    return ' '.join(tokens)

def calculate_cosine_similarity(str1, str2):
    vectorizer = CountVectorizer().fit_transform([str1, str2])
    vectors = vectorizer.toarray()
    return cosine_similarity([vectors[0]], [vectors[1]])[0][0]


from sentence_transformers import SentenceTransformer, util

def compute_embedding_similarity(paragraph1, paragraph2, model):
    lines_paragraph1 = paragraph1.split('.')
    lines_paragraph2 = paragraph2.split('.')

    # Initialize a list to store embeddings
    embeddings_paragraph1 = []
    embeddings_paragraph2 = []
    similarity_indx = []

    # Encode each line to obtain embeddings
    for i in range(min(len(lines_paragraph1), len(lines_paragraph2))):
        embedding_paragraph1 = model.encode(lines_paragraph1[i], convert_to_tensor=True)
        embedding_paragraph2 = model.encode(lines_paragraph2[i], convert_to_tensor=True)

        embeddings_paragraph1.append(embedding_paragraph1)
        embeddings_paragraph2.append(embedding_paragraph2)

        # Compute and print the cosine similarity for each pair of lines
        similarity = util.pytorch_cos_sim(embedding_paragraph1.unsqueeze(0), embedding_paragraph2.unsqueeze(0)).item()
        #print(f"Embedding similarity between the {i + 1} lines of paragraph1 and paragraph2: {similarity}")
        similarity_indx.append(similarity)

    return similarity_indx,embeddings_paragraph1, embeddings_paragraph2


# Initialize the model
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
def calculate_dice_similarity(str1, str2):
    # Tokenize the strings and remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens1 = [word.lower() for word in word_tokenize(str1) if word.isalnum() and word.lower() not in stop_words]
    tokens2 = [word.lower() for word in word_tokenize(str2) if word.isalnum() and word.lower() not in stop_words]

    intersection = len(set(tokens1).intersection(tokens2))
    avg_size = (len(tokens1) + len(tokens2)) / 2

    dice_similarity = 2 * intersection / (len(tokens1) + len(tokens2)) if avg_size != 0 else 0
    return dice_similarity

def calculate_tversky_similarity(str1, str2, alpha=1, beta=1, gamma=1):
    # Tokenize the strings and remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens1 = [word.lower() for word in word_tokenize(str1) if word.isalnum() and word.lower() not in stop_words]
    tokens2 = [word.lower() for word in word_tokenize(str2) if word.isalnum() and word.lower() not in stop_words]

    intersection = len(set(tokens1).intersection(tokens2))
    size_A_minus_B = len(set(tokens1) - set(tokens2))
    size_B_minus_A = len(set(tokens2) - set(tokens1))

    tversky_similarity = intersection / (alpha * intersection + beta * size_A_minus_B + gamma * size_B_minus_A)
    return tversky_similarity

def jaro_distance(str1, str2):
    # Initialize matching and transposition counts
    matches = 0
    transpositions = 0

    # Define the matching window size
    window_size = max(len(str1), len(str2)) // 2 - 1

    # Initialize lists to store matching characters
    str1_matches = [False] * len(str1)
    str2_matches = [False] * len(str2)

    # Count matches and transpositions
    for i in range(len(str1)):
        start = max(0, i - window_size)
        end = min(i + window_size + 1, len(str2))

        for j in range(start, end):
            if str1[i] == str2[j] and not str2_matches[j]:
                matches += 1
                str1_matches[i] = True
                str2_matches[j] = True
                break

    # Check for transpositions
    k = 0
    for i in range(len(str1)):
        if str1_matches[i]:
            while not str2_matches[k]:
                k += 1
            if str1[i] != str2[k]:
                transpositions += 1
            k += 1

    # Calculate Jaro similarity
    if matches == 0:
        return 0.0

    jaro_similarity = (matches / len(str1) + matches / len(str2) + (matches - transpositions / 2) / matches) / 3

    return jaro_similarity
```

```python
def get_param(para1,para2,no_lines):
  para1 = preprocessed(para1)
  para2 = preprocessed(para2)
  jaccard = []
  for i in range(no_lines):
    Lref = para1.split('.')[i]
    Lans = para2.split('.')[i]
    jaccard.append(calculate_jaccard_similarity(Lref,Lans))

  return jaccard

def get_paramH(para1,para2,no_lines):
  para1 = preprocessed(para1)
  para2 = preprocessed(para2)
  hamming_distance = []
  for i in range(no_lines):
    Href = para1.split('.')[i]
    Hans = para2.split('.')[i]
    hamming_distance.append(calculate_hamming_distance(Href,Hans))

  return hamming_distance

def get_paramEM(para1,para2,no_lines):
  para1 = preprocessed(para1)
  para2 = preprocessed(para2)
  embedding = []
  for i in range(no_lines):
    Eref = para1.split('.')[i]
    Eans = para2.split('.')[i]
    only_val,_,_ = compute_embedding_similarity(Eref,Eans,model)
    embedding.append(only_val)

  return embedding

def get_paramLV(para1,para2,no_lines):
  para1 = preprocessed(para1)
  para2 = preprocessed(para2)
  LevenshteinL= []
  for i in range(no_lines):
    LVref = para1.split('.')[i]
    LVans = para2.split('.')[i]
    LevenshteinL.append(Levenshtein.distance(LVref,LVans))

  return LevenshteinL

def get_paramCS(para1,para2,no_lines):
  para1 = preprocessed(para1)
  para2 = preprocessed(para2)
  cosine = []
  for i in range(no_lines):
    CSref = para1.split('.')[i]
    CSans = para2.split('.')[i]
    cosine.append(calculate_cosine_similarity(CSref,CSans))

  return cosine

def get_paramD(para1, para2, no_lines):
    para1 = preprocessed(para1)
    para2 = preprocessed(para2)
    dice = []
    for i in range(no_lines):
        Dref = para1.split('.')[i]
        Dans = para2.split('.')[i]
        dice.append(calculate_dice_similarity(Dref, Dans))

    return dice

def get_paramT(para1, para2, no_lines):
    para1 = preprocessed(para1)
    para2 = preprocessed(para2)
    tversky = []
    for i in range(no_lines):
        Tref = para1.split('.')[i]
        Tans = para2.split('.')[i]
        tversky.append(calculate_tversky_similarity(Tref, Tans))

    return tversky

def get_paramJR(para1, para2, no_lines):
    para1 = preprocessed(para1)
    para2 = preprocessed(para2)
    jaro = []
    for i in range(no_lines):
        JRref = para1.split('.')[i]
        JRans = para2.split('.')[i]
        jaro.append(jaro_distance(JRref, JRans))

    return jaro
```

```python
def cal_return_metrics(Ref,data_no):
  f_j = []
  f_c = []
  f_h = []
  f_l = []
  f_e = []
  f_d = []
  f_t = []
  f_jr =[]
  for i in range(1,701):
      strs = globals()["strs" + str(i)]

      jaccard_indx = get_param(Ref, strs, 5)
      f_j.append(jaccard_indx)

      cosine_indx = get_paramCS(Ref, strs, 5)
      f_c.append(cosine_indx)

      hamming_indx = get_paramH(Ref, strs, 5)
      f_h.append(hamming_indx)

      LevenshteinL_indx = get_paramLV(Ref, strs, 5)
      f_l.append(LevenshteinL_indx)

      embedding_indx = get_paramEM(Ref, strs, 5)
      f_e.append(np.array(embedding_indx).flatten())

      dice_indx = get_paramD(Ref, strs, 5)
      f_d.append(dice_indx)

      tversky_indx = get_paramT(Ref, strs, 5)
      f_t.append(np.array(tversky_indx))

      jaro_indx = get_paramJR(Ref, strs, 5)
      f_jr.append(jaro_indx)

  return f_j,f_c, f_h, f_l, f_e ,f_d ,f_t ,f_jr
```

```python
def cal_AccP(actual, predicted):
    acc_percntge = []
    for a, p in zip(actual, predicted):
        if a != 0:
            accuracy_percentage = (1 - abs(a - p) / a) * 100
        else:
            accuracy_percentage = 100 if a == p else 0
        acc_percntge.append(accuracy_percentage)
    return acc_percntge

acc_percntge = cal_AccP(score, pred_score)
for i, accuracy in enumerate(acc_percntge, 1):
    print(f"m{i}: {accuracy:.2f}%")
```
